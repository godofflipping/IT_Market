```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
# E = H
x = [((i ** 2) / 3 + i) / 3 for i in range(100)]
y = [((i ** 2) + i) / 3 for i in range(100)]
z = [(i) / 3 for i in range(100)]

fig = plt.figure(figsize=(16,8))
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z, 'blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# E = H/2
x = [(i * (5/3) - 0.77 * np.sin(0.865 * i) - 1.333 * np.cos(0.865 * i) + 1.333) / 3 for i in range(100)]
y = [((i ** 2) / 4 + i) / 3 for i in range(100)]
z = [(i) / 3 for i in range(100)]

fig = plt.figure(figsize=(16,8))
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z, 'blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# E = 2H
x = [(i * (-5/3) + 0.962 * np.sinh(1.73 * i) + 0.333 * np.cosh(1.73 * i) + 0.333) / 3 for i in range(100)]
y = [((i ** 2) * 2.5 + i) / 3 for i in range(100)]
z = [(i) / 3 for i in range(100)]

fig = plt.figure(figsize=(16,8))
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z, 'blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
