```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [i for i in range(10, 535, 35)]
y = [22, 26, 26, 25, 23, 23, 21, 21, 20, 19, 18, 16, 14, 14, 13.5]

fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
ax.set_xlabel('$U_{in}$, mV')
ax.set_ylabel('$K$')
ax.plot(x, y, color='blue')
plt.grid()
plt.show()
```

```python
x1 = [i for i in range(970, 1100, 10)]
y1 = [13, 14.5, 16.4, 18.3, 20.5, 23, 26.2, 23.2, 20.6, 18.6, 16.7, 14.5, 13.1]
x2 = [i for i in range(970, 1091, 1)]
y2 = [32 / np.sqrt(1 + (34 * (i - 1030) / 1030) ** 2) for i in range(970, 1091, 1)]

fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
ax.set_xlabel('$f$, kHz')
ax.set_ylabel('$K$')
ax.plot(x2, y2, color='blue')
ax.scatter(x1, y1, color='red')
plt.grid()
plt.show()
```
